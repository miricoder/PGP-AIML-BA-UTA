###  1)Working with Pandas (its more advanced than Lists/Arrarys in terms of analaysing data)



```python
model_list=['BMW','Mercedes','Audi']
print(model_list)
car_conditions_list=['NEW','USED','Damaged/Auction']
print(car_conditions_list)
quantity_list=[100,150,250]
print(quantity_list)
car_prices_list=[100,20,5]
print (car_prices_list)
```

```python
import pandas as pd 
car_sales_data=pd.DataFrame()
car_sales_data['ModelId'] = model_list
car_sales_data['Condition'] = car_conditions_list
car_sales_data['Quantity'] = quantity_list
car_sales_data['Price'] = car_prices_list

car_sales_data
```

```python
car_sales_data['Total_Amount'] = car_sales_data['Quantity'] * car_sales_data['Price']
car_sales_data
```

```python
car_sales_data.sample(n=2)
```
